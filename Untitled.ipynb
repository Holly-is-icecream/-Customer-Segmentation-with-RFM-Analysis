```python
import pandas as pd

# 读取桌面上的文件（Windows 路径注意转义符 \\）
file_path = r"C:\Users\陈浩林\Desktop\Online Retail.xlsx"
df = pd.read_excel(file_path)

# 查看前几行数据
df.head()
```

```python
# 查看数据结构
df.info()
```

```python
df.isnull().sum()
```

```python
# 去除缺失客户 ID 的行
df = df.dropna(subset=["CustomerID"])

# 去除退货订单（含 "C" 的 InvoiceNo）
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]

# 创建新列：订单总金额
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
df
```

```python
# 转换为 datetime 类型
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# 提取日期（不带时间）
df["Date"] = df["InvoiceDate"].dt.date
df
```

```python
import matplotlib.pyplot as plt

# 计算每日销售总额
daily_sales = df.groupby("Date")["TotalPrice"].sum()

# 画图
plt.figure(figsize=(14, 5))
daily_sales.plot()
plt.title("📈 Daily Sales Trend", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Total Sales (£)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_products = (
    df.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# 横向条形图
plt.figure(figsize=(10, 6))
top_products.sort_values().plot(kind="barh", color="skyblue")
plt.title("🏷 Top 10 Most Sold Products", fontsize=16)
plt.xlabel("Total Quantity Sold")
plt.tight_layout()
plt.show()
```

```python
# 国家销售额（不包含 UK 也可单独分析）
country_sales = (
    df.groupby("Country")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# 绘图
plt.figure(figsize=(10, 6))
country_sales.sort_values().plot(kind="barh", color="salmon")
plt.title("🌍 Top 10 Countries by Total Revenue", fontsize=16)
plt.xlabel("Total Revenue (£)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import datetime as dt

# 使用你已经清洗过的数据 df
# 确保 InvoiceDate 是 datetime 类型
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# 设定分析参考日（你可以用数据最后一天 +1）
snapshot_date = df["InvoiceDate"].max() + dt.timedelta(days=1)

# 每个客户的 R、F、M
rfm = df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (snapshot_date - x.max()).days,  # Recency
    "InvoiceNo": "nunique",                                   # Frequency
    "TotalPrice": "sum"                                       # Monetary
})

# 重命名列名
rfm.rename(columns={
    "InvoiceDate": "Recency",
    "InvoiceNo": "Frequency",
    "TotalPrice": "Monetary"
}, inplace=True)

# 观察 RFM 分布
rfm.describe()
```

```python
# 分位数打分（1到5）
rfm["R_score"] = pd.qcut(rfm["Recency"], 5, labels=[5, 4, 3, 2, 1])
rfm["F_score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
rfm["M_score"] = pd.qcut(rfm["Monetary"], 5, labels=[1, 2, 3, 4, 5])

# 合并为 RFM 分数（字符串）
rfm["RFM_Score"] = rfm["R_score"].astype(str) + rfm["F_score"].astype(str) + rfm["M_score"].astype(str)
```

```python
def rfm_level(df):
    if df["RFM_Score"] == "555":
        return "Top Customer"
    elif int(df["R_score"]) >= 4 and int(df["F_score"]) >= 4:
        return "Loyal"
    elif int(df["R_score"]) >= 4:
        return "Recent"
    elif int(df["F_score"]) >= 4:
        return "Frequent"
    else:
        return "Other"

rfm["Customer_Level"] = rfm.apply(rfm_level, axis=1)
rfm["Customer_Level"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(x="Customer_Level", data=rfm, order=rfm["Customer_Level"].value_counts().index, palette="pastel")
plt.title("📊 Customer Segmentation Based on RFM")
plt.xlabel("Customer Level")
plt.ylabel("Number of Customers")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
